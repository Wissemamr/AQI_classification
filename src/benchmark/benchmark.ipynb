```python
import matplotlib.pyplot as plt


def plot_metric(metric_name):
    plt.figure(figsize=(10, 6))
    plt.bar(df["Algorithms"], df[metric_name], color="plum", alpha=0.6)
    plt.xlabel("Algorithms")
    plt.ylabel(metric_name)
    plt.title(f"{metric_name} for the different classification algorithms")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.grid(axis="y")
    plt.show()


metrics = ["Accuracy", "Recall", "Precision", "F1-score"]
for metric in metrics:
    plot_metric(metric)
```

## Benchmarking results analysis : 
### 1. Accuracy : 
Accuracy refers to the proportion of correctly classified instances out of the total instances, which is calculated like this :
$$ \text{Accuracy} = \frac{\text{Number of correctly classified instances}}{\text{Total number of instances}}$$
$$ \text{Accuracy} = \frac{\text{TP + TN}}{\text{TP+TN+FP+FN}}$$

### 2. Recall :
Recall (also known as sensitivity or true positive rate) measures the proportion of actual positive instances that are correctly identified by the model.
$$ \text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}}$$



### 3. Precision :
Precision measures the proportion of correctly identified positive instances out of all instances that were classified as positive by the model.
$$\text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}$$

### 4. F1-Score : 
The F1 score is a metric that combines both precision and recall into a single value. It's the harmonic mean of precision and recall and provides a balance between the two.
$$F1\text{-} \text{score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$$

## Best Algorithms
As we can see from the plots, the models that performed the best in terms of the classification metrics we chose : **Accuracy, Recall, Precison, F1-score** are :
- Decision trees
- Random Forests
- Gradient Boosting Machine
- XGBoost
- SVM
- AdaBoost

- **Decision Trees**: Decision trees are easy to interpret and understand, making them suitable for exploratory analysis. They can handle both numerical and categorical data and are robust to outliers. Decision trees are capable of capturing non-linear relationships between features and the target variable, making them effective for a wide range of classification tasks, which was the case for out data.
- **Random Forests**: Random Forests are an ensemble learning method that builds multiple decision trees and combines their predictions through voting or averaging. They are robust against overfitting and tend to generalize well to unseen data. By constructing multiple trees and aggregating their predictions, Random Forests can capture complex relationships in the data and handle high-dimensional feature spaces effectively.
- **Gradient Boosting Machine (GBM)**: Gradient Boosting is another ensemble learning technique that builds a sequence of decision trees, where each subsequent tree corrects the errors made by the previous ones. GBM sequentially fits the residuals of the previous trees, which allows it to gradually improve the model's performance. This iterative process often leads to highly accurate predictions, especially when combined with shallow trees to prevent overfitting.
- **XGBoost**: Extreme Gradient Boosting (XGBoost) is an optimized implementation of gradient boosting with several additional features, such as regularization, parallel computation, and support for missing values. XGBoost often outperforms traditional gradient boosting methods due to its efficiency and scalability. It incorporates advanced techniques to control overfitting and improve generalization performance, making it a popular choice for various classification tasks.
- **Support Vector Machines (SVM)**: SVM is a powerful supervised learning algorithm that is effective for both classification and regression tasks. By maximizing the margin between the classes, SVM can achieve good generalization performance, even in high-dimensional spaces. SVM is also effective in cases where the data is not linearly separable by using kernel tricks to map the data into a higher-dimensional space.

## Notes : 
- If we shed light on the **Naive Bayes approach**, we notice that it has performed rather poorly compared to the other algortihms that gave pretty good results. THis is mainly due to the assumption we make when using naive Bayes which is that the features are independant. This assumption implies that all features in the dataset are independent of each other given the class label. In reality, this assumption rarely holds true for complex real-world data. When features are correlated or dependent on each other, Naive Bayes may fail to capture these relationships accurately, leading to suboptimal performance
